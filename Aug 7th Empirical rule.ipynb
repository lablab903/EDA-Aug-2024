**Emperical Rule**

- (mean-1 std) to (mean+1std) : 68%
- (mean-2 std) to (mean+2std) : 95%
- (mean-3 std) to (mean+2std) : 98%

```python
# step-1: calculate the mean of prevailing wage

# step-2: calculate the std of prevailig wage

# step-3: lb=mean-1*std                                      lb is lower bound                         

# step-4: ub=mean+1*std                                      ub is upper bound

# step-5: con1= data<ub                                     condition 1

# step-6: con2= data>lb                                     condition 2

# step-7: con3= con1 and con2                               condition 3

# step-8: data[con3]

# step-9: compare with 68% 25480 count
```

```python
# PART 1: to find the prevailing wage between lb and ub values: 

mu = visa_df['prevailing_wage'].mean()

print("\n\n")

print(" the mean of prevailing wage is =",round(mu,2))

sigma = visa_df['prevailing_wage'].std()

print("\n the std deviation of prevailing wage is = ",round(sigma,2))

lb = mu-1*sigma                                                                     # formula for lb

print(" \nthe lower bound value is = ",round(lb,2))

ub = mu+1*sigma                                                                     # formula for lb

print("\n the upper bound value is =",round(ub,2))

print("\n\n")

##########################################################################################

```

```python
# PART 2: prevailing wage between lb and ub values of 21639 and 127271

con1=visa_df['prevailing_wage'] > lb

con2=visa_df['prevailing_wage'] < ub

con3= con1 & con2                               # condition 1 and 2 is True

visa_df[con3]
```
