```python
# Continent column ====> continent_df ====> plt.bar
# continent column directly ===> visadf

#  seaborn ie sns  with plt (matplotlib)

import seaborn as sns

plt.figure(figsize=(8,5))                        # 8 units X-axis, 5units y-axis ;                     plt is from matplotlib pkg

order_continents=['Asia','Europe','North America','South America','Africa','Oceania']           # giving continent order manually

sns.countplot(data=visa_df,                                                                        # USE countplot() method of sns
             x='continent',
             order=order_continents)

plt.title('Bar chart')                                                                            # Using plt pkg to make graph title         

plt.savefig('Continents_bar_chart_using_seaborn.jpg')                                          # Using plt pkg to save the pic of graph

plt.show()                                                                              # Using plt pkg to display the graph
```

```python
#  create bar chart for all 8 cat column using for loop with 'i'
# the 8 cols are : continent, education of employee, has job experinnce, requires job training , 
# region of employment, unit of wage, full time position, case status


import seaborn as sns                                                 # seaborn will give COLOR bar charts

for i in cat_columns[1:]:
    
    plt.figure(figsize=(8,5))                                             # 8 units X-axis, 5units y-axis
    
    order_continents=visa_df[i].value_counts().keys()
    
    sns.countplot(data=visa_df,
                 x=i,
                 order=order_continents)
    
    plt.title('Bar chart')
    
    plt.savefig(f'{new_dir}\\{i}_seaborn.jpg')                          # use f string 
    
    plt.show()                                                                  # show should be at last
```

```python
import matplotlib.pyplot as plt                                     # MATPLOTLIB gives only BLUE COLOR bar graph columns

for i in dfs:
    
    df=pd.read_csv(i)
    
    plt.bar('Lables',                                                # Cat column  X-axis
               'Count',                                               # Num column: Height Y-axis
                data=df)
    
    plt.title("Bar chart")                                            # Title of the graph
    
    plt.xlabel('Label')                                                # X-axis name 
    
    plt.ylabel('Count')                                                  # Y-axis name
    
    plt.savefig(f'{new_dir}\\{i}_matplotlib.jpg')
    
    plt.show()                                                    # Show will avoid the writeup and it should be at last
```

```python
# for 1 column to generate 'education_of_employee' bar chart

cdf=visa_df['education_of_employee'].value_counts()

keys=cdf.keys()

values=cdf.values

cols=['Lables','Count']

df=pd.DataFrame(zip(keys,values),
             columns=cols)

df.to_csv('education_of_employee.csv',index=False)
```
