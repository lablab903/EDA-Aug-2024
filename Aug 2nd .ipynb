```python
# separating categorical coumn from numerical cols using 'object' asit is categoricaltype

cat_columns=visa_df.select_dtypes(include='object').columns
num_columns=visa_df.select_dtypes(exclude='object').columns
cat_columns

```

```python
num_columns                                                    # numerical columns in visadataset.csv file...3 cols
```

```python
cat_columns[1:]                                                  # 8 categorical columns
```

```python
# subplots (3,3,0)  or (3,3,1) = (3,3,1)
# for i in range , what is the start value of i = 0
```

```python
# USING FOR LOOP TO to generate 8 graphs  USING MATPLOTLIB pkg and SEABORN PKG since their are 8 categorical columns

cols=cat_columns[1:]

plt.figure(figsize=(16,14))

for i in range(len(cols)):
    
    plt.subplot(3,3,i+1)
    sns.countplot(x=cols[i],data=visa_df)

```
